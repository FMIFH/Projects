### prep

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split

import xgboost as xgb

from sklearn.metrics import accuracy_score, roc_auc_score
from sklearn.model_selection import GridSearchCV
```

```python
df = df2
```

```python
X = df.drop('target',axis=1)
y = df.target
```

```python
X.shape
```

```python
y.shape
```

```python
numeric_pipeline = Pipeline(
    steps=[("scale", StandardScaler())]
)
```

```python
full_processor = ColumnTransformer(
    transformers=[
        ("numeric", numeric_pipeline, X.columns),
    ]
)
```

```python
X_processed = full_processor.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_processed, y,test_size=0.3, stratify=y)
```

```python
X_processed.shape
```

```python
param_grid = {
    "max_depth": [1, 2, 3, 4, 5, 7],
    "learning_rate": [0.5, 0.1, 0.05, 0.01],
    "gamma": [0, 0.1, 0.25, 1, 2],
    "reg_lambda": [0, 0.1, 1, 10],
    "scale_pos_weight": [1, 3, 5, 10],
    "subsample": [0.8],
    "colsample_bytree": [0.5],
}
```

```python
xgb_cl = xgb.XGBClassifier()
grid_cv = GridSearchCV(xgb_cl, param_grid, n_jobs=-1, cv=5, scoring="roc_auc")

_ = grid_cv.fit(X_train, y_train)
```

```python
grid_cv.best_score_
```

```python
grid_cv.best_params_
```

```python
final_cl = xgb.XGBClassifier(
    **grid_cv.best_params_,
    objective="binary:logistic",
)

_ = final_cl.fit(X_train, y_train)
```

```python
predictions = final_cl.predict(X_test)
```

```python
roc_auc = roc_auc_score(y_test,predictions)
roc_auc
```

```python
accuracy = accuracy_score(y_test, predictions)
accuracy
```

```python
feature_important = final_cl.feature_importances_
```

```python
feature_important
```

```python
keys = X.columns.values
values = feature_important

data = pd.DataFrame(data=values, index=keys, columns=["score"]).sort_values(by="score", ascending=False)
data.nlargest(10, columns="score").sort_values('score', ascending=True).plot(kind='barh', figsize = (20,10))
```
