```python
from serpapi import GoogleSearch
import pandas as pd
import datetime
```

```python
key_words = ["remote","Python","SQL","Tableau","ML","Machine Learning","Deep Learning","Java","PowerBI","Tableau","Excel", "Power BI", "Office"]
level = ["senior","entry","internship","full time"]
```

```python
search_term = "data analyst"
search_location = "Portugal"

for num in range(0,45):
    start = num * 10
    print(start)
    params = {
        "api_key": "3a3223763be01f03177328d7356ff5bf82e36164e26877f7f0f732b31ff64347", #Fill in with your API key from SerpApi
        "device": "desktop",
        "engine": "google_jobs",
        "google_domain": "google.com",
        "q": search_term,
        "hl": "en",
        "gl": "us",
        "location": search_location,
        "chips": "date_posted:today",
        "start": start,
    }

    search = GoogleSearch(params)
    results = search.get_dict()

    # check if the last search page (i.e., no results)
    try:
        if results['error'] == "Google hasn't returned any results for this query.":
            break
    except KeyError:
        print(f"Getting SerpAPI data for page: {start}")
    else:
        continue

    # create dataframe of 10 pulled results
    jobs = results['jobs_results']
    jobs = pd.DataFrame(jobs)
    jobs = pd.concat([pd.DataFrame(jobs), 
                    pd.json_normalize(jobs['detected_extensions'])], 
                    axis=1).drop('detected_extensions', 1)
    jobs['date_time'] = datetime.datetime.utcnow()
    jobs.drop(["thumbnail"],axis=1)
    jobs["Technologies"] = [[w for w in key_words if w.upper() in d.upper()] for d in jobs.description]
    jobs["Level"] = [[l for l in level if l.upper() in d.upper()] for d in jobs.description]

    # concat dataframe
    if start == 0:
        jobs_all = jobs
    else:
        jobs_all = pd.concat([jobs_all, jobs],ignore_index=True)

    jobs_all['search_term'] = search_term
    jobs_all['search_location'] = search_location
```
