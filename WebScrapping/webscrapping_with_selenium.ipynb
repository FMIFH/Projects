```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
driver = webdriver.Chrome("C:\\Users\\franc\\Downloads\\chromedriver_win32\\chromedriver.exe")
```

```python
LIST = 'https://www.imdb.com/list/ls009796553/?sort=list_order,asc&st_dt=&mode=detail&page='
```

```python
def make_movie_object(name,year,certificate,rate,runtime,genre,description,directors,stars,gross):
    director = []
    star = []
    for d in directors:
        director.append(d.strip())
    for s in stars:
        star.append(s.strip())
    movie_dict = {
        'name' : name,
        'year' : year,
        'certificate' : certificate,
        'rating' : rate,
        'runtime (min)' : runtime,
        'genre' : genre,
        'description' : description,
        'directors' : director,
        'stars' : star,
        'gross' : gross,
    }
    return movie_dict
```

```python
movies = []
for i in range(1,8):
    driver.get(LIST+str(i))
    content = driver.page_source
    soup = BeautifulSoup(content)
    for a in soup.findAll('div', attrs={'class':'lister-item-content'}):
        try:
            header = a.find('h3',attrs={'class':'lister-item-header'})
            name = header.find('a').text
            year = int(header.find('span',attrs={'class':"lister-item-year text-muted unbold"}).text.replace(' Video',"").replace("(I) ", "").replace("(", "").replace(")", ""))
            subheader = a.find('p',attrs={'class':'text-muted text-small'})
            certificate = subheader.find('span',attrs={'class':'certificate'}).text

            runtime = int(subheader.find('span',attrs={'class':'runtime'}).text.replace(" min", ""))
            genre = subheader.find('span',attrs={'class':'genre'}).text.replace(" ", "").replace("\n", "").split(',')
            rating = float(a.find('span',attrs={'class':'ipl-rating-star__rating'}).text)
            description = a.findAll('p')[1].text.strip()
            
            info = a.findAll('p',attrs={'class':'text-muted text-small'})
            crew = info[1].text
            directors = crew.split('|')[0].replace("Director:\n","").replace("Directors:\n","").replace("\n", "").split(',')
            stars = crew.split('|')[1].replace("Stars:\n","").replace("\n", "").strip().split(',')
            gross = int(info[2].findAll('span',attrs={'name':"nv"})[1]['data-value'].replace(',', ''))
            movies.append(make_movie_object(name,year,certificate,rating,runtime,genre,description,directors,stars,gross))
        except:
            print(f'Failed for {name} page {i}')
            continue
driver.close()
```

```python
df = pd.DataFrame(movies)
```
