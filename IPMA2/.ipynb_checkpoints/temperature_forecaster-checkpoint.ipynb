```python
import os

import matplotlib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from prophet import Prophet
```

```python
BOUNDARY_DATE = '01-01-2000'
```

```python
df = pd.read_csv("temp-Lisboa_Geof√≠sico-hom.csv")
```

```python
df.info()
```

```python
df.dropna(inplace=True)
df.drop(["tmin","tmax"], inplace=True,axis=1)
```

```python
df.columns = ["ds","y"]
```

```python
df["ds"] = df["ds"].apply(pd.Timestamp)
```

```python
plt.figure(figsize=(16, 9))
sns.lineplot(data=df, x='ds', y='y')
plt.show()
```

```python
train_df = df.where(df.ds < BOUNDARY_DATE).dropna()
test_df = df.where(df.ds >= BOUNDARY_DATE).dropna()
```

```python
m = Prophet(seasonality_mode='multiplicative')
m.fit(train_df)
```

```python
future_dataframe = m.make_future_dataframe(periods=len(test_df)+1,freq='m')
dataframe2025 = m.make_future_dataframe(periods=35*12,freq='m')
forecast = m.predict(future_dataframe)
forecast2025 = m.predict(dataframe2025)
```

```python
# Split forecast into train and test parts
forecast_train = forecast.where(forecast.ds < BOUNDARY_DATE).dropna()
forecast_test = forecast.where(forecast.ds >= BOUNDARY_DATE).dropna()
```

```python
test_df, forecast_test = test_df.align(forecast_test, axis=0)
anomalies = test_df[(test_df["y"] > forecast_test["yhat_upper"]) | (test_df["y"] < forecast_test["yhat_lower"])]
```

```python
plt.figure(figsize=(16, 9))

sns.lineplot(data=test_df, x='ds', y='y')
sns.lineplot(data=forecast_test, x='ds', y='yhat', alpha=0.7)
plt.fill_between(x='ds',y1='yhat_lower',y2='yhat_upper',data = forecast_test, alpha = 0.4)
sns.scatterplot(data=anomalies, x='ds', y='y',color= 'red')

plt.show()
```
