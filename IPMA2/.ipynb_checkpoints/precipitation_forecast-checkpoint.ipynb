```python
import os

import matplotlib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from prophet import Prophet
```

```python
sns.set()
```

```python
FILE_PATH = "diario/PRECIP_PT/"
DATA_FILES = [os.path.join(FILE_PATH,i) for i in next(os.walk(FILE_PATH))[2]]
```

```python
date = pd.Series(dtype=np.int64)
precip = pd.Series(dtype=np.float64)
for f in DATA_FILES:
    data = pd.read_fwf(f, sep="\t", header=None)
    #select lisbon data
    data = data[((data[0]==38.8) & (data[1]==-9.2))|(data[0]>1000)].reset_index()
    date = pd.concat([date,data[0].iloc[::2].astype(np.int64)],ignore_index=True)
    precip = pd.concat([precip,data[2].iloc[1::2]],ignore_index=True)
```

```python
d = {'ds':date,'y':precip}
data = pd.DataFrame(data=d)
data["ds"] = pd.to_datetime(data["ds"],format="%Y%m%d")
data.dropna(inplace=True)
data.sort_values(by=['ds'],inplace=True)
```

```python
data.info()
```

```python
plt.figure(figsize=(15,8))
sns.lineplot(data=data,x='ds', y='y')
plt.show()
```

```python
m = Prophet()
```

```python
model = m.fit(data)
```

```python
future = m.make_future_dataframe(periods=20*365)
forecast = m.predict(future)
```

```python
m.plot(forecast)
```

```python
fig2 = m.plot_components(forecast)
```
